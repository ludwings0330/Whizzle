```python
from selenium import webdriver
import urllib.request
import time
import os
import pandas as pd
```

```python
if not os.path.isdir("whisky_images/"):
    os.makedirs("whisky_images/")
```

```python
df = pd.read_csv('C:/Users/SSAFY/Downloads/whisky.csv')
df = df['name']
df
```

```python
driver = webdriver.Chrome()
```

```python
for i, whisky_name in enumerate(df):
    whisky_name = whisky_name.replace('#', '')
    print(i, whisky_name)

    if i < 135:
        continue
    # Navigate to Google Images
    driver.get(f"https://www.google.com/search?q={whisky_name}%20PNG&tbm=isch")
    images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
    image = images[0]
    image.click()

    imgUrl = driver.find_element_by_xpath('//*[@id="Sva75c"]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div[1]/div[2]/div[2]/div/a/img').get_attribute('src')
    urllib.request.urlretrieve(imgUrl, "whisky_images/NO_" + str(i) + ".png")

    time.sleep(2)
        
```
